# Starting OpenCV

The first thing you need to do to use OpenCV is to import the package. Notice that even though the version of Opencv is 3.x.x/4.x.x you should import it by using <tt class="docutils literal"><span class="pre">import cv2</span></tt>


```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
print ("OpenCV Version: %s " % cv2.__version__)#Check the version of opencv
```

```python
# Load an image using opencv
import cv2
from matplotlib import pyplot as plt
my_image = cv2.imread("dog.jpg")
plt.figure(figsize=(4,4))
plt.imshow(my_image)
plt.show()
```

Notice that the color of the image is not correct. The reason is that <tt class="docutils literal"><span class="pre">matplotlib</span></tt> expects an RGB image but <tt class="docutils literal"><span class="pre">opencv</span></tt> image is a BGR image. Use the <tt class="docutils literal"><span class="pre">cvtColor</span></tt> in <tt class="docutils literal"><span class="pre">opencv</span></tt> to convert from BGR to RGB.

```python
my_rgb_image = cv2.cvtColor(my_image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(4,4))
plt.imshow(my_rgb_image)
plt.show()
```
