```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
cameraman = cv2.imread("camera.jpeg", cv2.IMREAD_GRAYSCALE)
```

```python
cameraman.shape
```

```python
robert = np.array([[1,0],[0,-1]])
robert2 = np.array([[0,-1],[1,0]])
```

```python
result = cv2.filter2D(cameraman,ddepth=0,kernel=robert)
```

```python
plt.imshow(result, cmap="gray")
```

```python
result2 = cv2.filter2D(cameraman, ddepth=0, kernel=robert2)
```

```python
plt.imshow(result2, cmap="gray")
```

```python
sobelx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobely = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
```

```python
sobx = cv2.filter2D(cameraman, ddepth=0, kernel = sobelx)
soby = cv2.filter2D(cameraman, ddepth=0, kernel = sobely)
```

```python
plt.subplot(1,2,1)
plt.title("Sobel X")
plt.imshow(sobx, cmap="gray")

plt.subplot(1,2,2)
plt.title("Sobel Y")
plt.imshow(soby,cmap="gray")
```

```python
laplacian = np.array([[0,1,0],[1,-4,1],[0,1,0]])
```

```python
lap = cv2.filter2D(cameraman, ddepth=0,kernel=laplacian)
```

```python
plt.imshow(lap,cmap="gray")
```
