## Import Libraries

```python
# Import libraries
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

## Read an image

```python
# Load the image
image = cv2.imread("Images/dog.jpg")

# Display the original image
plt.imshow(image[:,:,::-1])
```

## Draw a Line

We can use cv2.line function for this task.

Function Syntax: cv.line(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) ->	img

Parameters:

    img	Image.
    pt1	First point of the line segment.
    pt2	Second point of the line segment.
    color	Line color.
    thickness	Line thickness.
    lineType	Type of the line. See LineTypes.
    shift	Number of fractional bits in the point coordinates.



```python
# Draw a line
imageLine = image.copy()

# The line starts from (200,80) and ends at (280,80)
# The color of the line is RED (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.line(imageLine, (120, 80), (270, 80), (0, 255, 0), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageLine[:,:,::-1])
```

```python
cv2.line(imageLine, (120, 200), (270, 200), (0, 255, 0), thickness=5, lineType=cv2.LINE_AA);
cv2.line(imageLine, (120, 80), (120, 200), (0, 255, 0), thickness=5, lineType=cv2.LINE_AA);
cv2.line(imageLine, (270, 80), (270, 200), (0, 255, 0), thickness=5, lineType=cv2.LINE_AA);
# Display the image
plt.imshow(imageLine[:,:,::-1])
```

## Drawing a rectange

To draw a rectangle, you need top-left corner and bottom-right corner of rectangle. This time we will draw a red rectangle on the dog head.

```python
# Draw a rectangle
imageRec = image.copy()

# The top corner is at (120,80) and bottom corner at (270,200)
# The color of the line is RED (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.rectangle(imageRec, (120, 80), (270, 200), (0, 0, 255), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageRec[:,:,::-1])
```

## Draw a Circle

We can use cv2.circle function to draw a circle on an image.

```python
# Draw a Circle
imageCircle = image.copy()

# The center is at (200,150) and the radius is 75
# The color of the line is BLUE (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.circle(imageCircle, (200, 150), 75, (255, 0, 0), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageCircle[:,:,::-1])
```

## Drawing an Ellipse


```python
# Draw an Ellipse
imageEli = image.copy()

# The center is at (200,150) and the radius is 75
# The color of the line is BLUE (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.ellipse(imageEli, (200, 150), (150,75), 0, 0, 360, (255, 0, 0), thickness=5, lineType=cv2.LINE_AA);
cv2.ellipse(imageEli, (200, 150), (150,75), 90, 0, 360, (0, 0, 255), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageEli[:,:,::-1])
```

## Adding Texts to Images

```python
imageText = image.copy()

cv2.putText(imageText,'Hello World!!',(10,50), cv2.FONT_HERSHEY_SIMPLEX, 1,(255,255,255),2,cv2.LINE_AA);

# Display Image with Text
plt.imshow(imageText[:,:,::-1])
```
